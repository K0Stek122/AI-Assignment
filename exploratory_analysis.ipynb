```python
labels = ['Negative', 'Positive']
negative_review_count = len(df[df["class"] == 0])
positive_review_count = len(df[df["class"] == 1])
sizes = [negative_review_count, positive_review_count]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python
reviews = ''.join(list(df["review"].values)) #turn all reviews into one long string
reviews = reviews.lower().rstrip() # Strip reviews of leading \n's
stripped_reviews = ''.join(ch for ch in reviews if ch.isalnum()) #remove any non-alphanumeric characters
stripped_reviews = reviews.replace("!", "").replace(r'"', '').replace('-', '').replace(".", "").split(" ") # Replace any other unwanted characters
st = set(stripped_reviews) # Turn into a set to get unique words
len(st) # Print unique words. These will be the features of the AI.
```
